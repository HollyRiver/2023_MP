### **C. 결측치 처리**
> 수치형은 수치형끼리, 범주형은 범주형끼리 처리하자.

````python
train_num = df_train.select_dtypes(include = 'number') ## int와 float 모두 걸러진다.
train_cat = df_train.select_dtypes(exclude = 'number')
````

> 이런 식으로 처리하면 되는데, 데이터 여럿을 처리하려면 귀찮으니 함수를 정의해보자.

````python
def impute_missing(df) :
    df_imputed = df.copy()
    df_num = df.select_dtypes(include = 'number')
    df_cat = df.select_dtypes(exclude = 'number')
    
    df_imputed[df_num.columns] = sklearn.impute.SimpleImputer(strategy = 'mean').fit_transform(df_num)
    df_imputed[df_cat.columns] = sklearn.impute.SimpleImputer(strategy = 'most_frequent').fit_transform(df_cat)
    
    return df_imputed

msno.matrix(impute_missing(df_train))
````

> 결측치가 발견되지 않는 모습이다.

## 4. 분석

```'Survived'``` : 반응변수

````python
X = pd.get_dummies(impute_missing(df_train).drop(['Survived'], axis = 1))  ## drop()는 기본적으로 열을 드랍
y = impute_missing(df_train)[['Survived']]
````

> Name, Ticket, Cabin의 경우에는 중복되지 않은 변수들의 개수가 너무 많다.
>
> 따라서 배제하도록 한다.
> * set(array) : 중복되지 않는 데이터만 산출

````python
X = pd.get_dummies(impute_missing(df_train).drop(['Survived', 'Name', 'Ticket', 'Cabin'], axis = 1))
y = df_train.Survived
````

````python
X
````

### **B. predictor 생성**

````python
predictr = sklearn.linear_model.LogisticRegression()
````

### **C. 학습**

````python
predictr.fit(X, y)
````

### **D. 예측**

````python
predictr.predict(X)
````

### **E. 평가**

````python
predictr.score(X,y)
````
