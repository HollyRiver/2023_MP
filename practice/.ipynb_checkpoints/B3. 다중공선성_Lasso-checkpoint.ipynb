> `Lasso`를 이용해 다중공선성을 해소해보자.

## 1. 라이브러리 imports

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import sklearn.linear_model
```

## 2. Data

```python
df = pd.read_csv("https://raw.githubusercontent.com/guebin/MP2023/main/posts/employment_multicollinearity.csv")
np.random.seed(43052)
df['employment_score'] = df.gpa * 1.0 + df.toeic* 1/100 + np.random.randn(500)

df
```

## 3. True(Oracle)

```python
## step1 
df_train, df_test = sklearn.model_selection.train_test_split(df,test_size=0.3,random_state=42)
X = df_train.loc[:,'gpa':'toeic']
y = df_train[['employment_score']]
XX = df_test.loc[:,'gpa':'toeic']
yy = df_test[['employment_score']]
## step2 
predictr = sklearn.linear_model.LinearRegression()
## step3
predictr.fit(X,y)
## step4 : pass 
```

```python
print(f'train_score:\t{predictr.score(X,y):.4f}')
print(f'test_score:\t{predictr.score(XX,yy):.4f}')
```

## 4. Lasso
\- `Lasso`를 이용

```python
## step1 
df_train, df_test = sklearn.model_selection.train_test_split(df,test_size=0.3,random_state=42)
X = df_train.loc[:,'gpa':'toeic499']
y = df_train[['employment_score']]
XX = df_test.loc[:,'gpa':'toeic499']
yy = df_test[['employment_score']]
## step2 
predictr = sklearn.linear_model.Lasso()  ## alpha default : 1.0
## step3
predictr.fit(X,y)
## step4 : pass 
```

```python
print(f'train_score:\t {predictr.score(X,y):.4f}')
print(f'test_score:\t {predictr.score(XX,yy):.4f}')
```

> `alpha`를 `defalut`로 두었음에도 굉장히 긍정적인 결과가 발생...

## 5. Lasso는 왜 결과를 좋게 만들까?

### **A. 정확한 설명**
---
\- 어려움...

대충 내가 검색해서 얻은 결과는

$$MSE + \alpha L_1{-}norm$$
$$=\frac{1}{n}\sum_{i = 1}^{n}{(y_i - \hat{y}_i)^2} + \alpha\sum_{i = 1}^$$

### **B. 직관적 설명(엄밀하지 않은 설명)**
---
\- 몇 개의 `toeic coef`들로 쉽게 0.01을 만들게 해서는 안된다.

* 아이디어 1 : 0.01을 동일한 값으로 균등하게 배분 : Ridge, L2-penalty
* 아이디어 2 : 아주 적은 숫자의 coef만을 살려두고 나머지 coef 값은 0으로 강제한다 : Lasso, L1-penalty

\- 계수값이 0이라는 의미 : 해당 변수를 제거한 것과 같은 효과

> `y` ~ `toeic` + `gpa`가 True이지만, `y` ~ `toeic0` + `gpa`로 적합해도 괜찮잖아?

\- 진짜 학습된 계수값이 대부분 0인지 확인해보자.

```python
plt.plot(predictr.coef_[1:])
plt.show()
```

> 어떤 개체가 0이 될지는 모르겠지만, 대부분이 0이 되어버림.

### **$\alpha$에 따른 변화 관찰**
---
\- 여러 개의 predictor 학습

```python
## step1 
df_train, df_test = sklearn.model_selection.train_test_split(df,test_size=0.3,random_state=42)
X = df_train.loc[:,'gpa':'toeic499']
y = df_train[['employment_score']]
XX = df_test.loc[:,'gpa':'toeic499']
yy = df_test[['employment_score']]
## step2 
alphas = np.linspace(0,2,100)
predictrs = [sklearn.linear_model.Lasso(alpha=alpha) for alpha in alphas]
## step3
for predictr in predictrs: 
    predictr.fit(X,y);
## step4 : pass
```

```python
plt.plot(predictrs[0].coef_[1:], label=r'$\alpha={}$'.format(predictrs[0].alpha))
plt.plot(predictrs[50].coef_[1:], label=r'$\alpha={}$'.format(predictrs[50].alpha))
plt.plot(predictrs[-1].coef_[1:], label=r'$\alpha={}$'.format(predictrs[-1].alpha))
plt.legend()
plt.show()
```

> $\alpha$의 값이 커질수록 coef들의 분산이 작아짐(아마도...)

```python
print(f'alpha={predictrs[0].alpha:.4f}\tsum(toeic_coef)={predictrs[0].coef_[1:].sum()}')
print(f'alpha={predictrs[50].alpha:.4f}\tsum(toeic_coef)={predictrs[50].coef_[1:].sum()}')
print(f'alpha={predictrs[-1].alpha:.4f}\tsum(toeic_coef)={predictrs[-1].coef_[1:].sum()}')
```

> predictor들의 toeic 계수합은 여전히 0.01 근처

\- number of non-zero coefs를 시각화하면...

```python
non_zero_coefs = [(abs(predictr.coef_[1:]) > 0).sum() for predictr in predictrs]
##non_zero_coefs = [(predictr.coef_[1:] != 0).sum() for predictr in predictrs] 와 동일
```

```python
plt.plot(alphas, non_zero_coefs)
plt.show()
```

### **D. coef를 0으로 만드는 수학적 장치**
---
\- Ridge(복습) : coef의 값들을 가중치에 따라 엔빵하는 수학적 장치

*  패널티 : 유사토익들의 계수값을 제곱한 뒤 합치고(제곱합)(L2-norm을 구하고), 그 값이 0에서 떨어져 있을수록 패널티를 줄 거야!

\- Lasso : coef의 값들을 대부분 0으로 만드는 수학적 장치

* 패널티 : 유사토익들의 계수값의 절대값을 구한 뒤에 합치고(=L1-norm을 구하고),그 값이 0에서 떨어져 있을수록 패널티를 줄거야!

\- 사실 L1, L2 패널티에 따라서 이러한 결과가 나오는 것은 이해하기 어렵다.(취업/대학원 진학 시 단골질문 중 하나)
